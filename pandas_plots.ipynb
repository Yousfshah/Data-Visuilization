# Data Visulization Using Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

# **1.Bar Chart**

```python
# Create a dict with a sample data
data={'Fruits':['Apple','Banana','Mango'],
     'Sales':[100,150,80]}

# Create a dataframe
df=pd.DataFrame(data)

# Plot a dataframe as a Bar Chart
c=['b', 'g', 'c']
df.plot(kind='bar', x='Fruits', y='Sales', legend=True, figsize=(7,5), 
        color=c, edgecolor='black', linewidth=2)

# Add labels and title
plt.xlabel('Fruits', fontsize=14, fontweight='bold')
plt.ylabel('Revenue', fontsize=14, fontweight='bold')
plt.title('Fruits Sales during the period', fontsize=18, fontweight='bold')

# Add grid lines
plt.grid(axis='y', linestyle='-.', alpha=0.4)

# Save plot to file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Pandas_Plots/Bar_Chart.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

# **2. Pie Chart** 


```python
# Create a dict with a sample data
data={'Fruits':['Apple','Banana','Mango'],
     'Sales':[100,150,80]}

# Create a dataframe
df=pd.DataFrame(data)

# Plot the dataframe as Pie Chart
colors = {'#012602','#000000','#00012e'}
df.set_index('Fruits')['Sales'].plot(kind='pie', autopct='%1.1f%%',startangle=90,colors=colors,textprops={'color':'green'})

# set the title of the plot
plt.title('Sales distribution of different Fruits',fontsize=16)

# Save plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Pandas_Plots/Pie_Plot.png', dpi=300, bbox_inches='tight')

# show plot
plt.ylabel('')  # remove y-axis label
plt.show()
```

# **3.Histogram**

```python
# create a dictionary with sample data
data = {'Sales': [30,50,120,150,200,300,100, 150, 80, 120, 140, 130,
                  50,50,30,20,10,50,35,45,55,50,55,65,56,58,58,54,55,50,5,
                  110, 145, 105, 90, 115, 125, 135, 85, 95]}

# create a pandas dataframe from the dictionary
df = pd.DataFrame(data)

# plot the dataframe as a Histogram
df.plot(kind='hist',bins=5,edgecolor='black',color='green')

# Adding grid
plt.grid(True , linestyle=':' , linewidth=0.5,color='green')

# Adding histogram title and labels
plt.title('Distribution of Sales',fontsize=16)
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')

# Adjusting the X-Axis Ticks
plt.xticks(range(0,350,50))
plt.xticks(rotation=45)

# Adjusting the Y-Axis Ticks
plt.yticks(range(0,23,3))
plt.yticks(rotation=45)

# save the plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Pandas_Plots/Histogram.png', dpi=300, bbox_inches='tight')

# show plot
plt.show()
```

# **4.Scatter Plot**

```python
# create a dictionary with sample data
data = {
    'Sales': [100, 150, 80, 120, 140, 130, 110, 145, 105, 90],
    'Profit': [50, 70, 40, 60, 75, 65, 55, 72, 52, 45]
}

# create a pandas dataframe from the dictionary
df = pd.DataFrame(data)

# plot the dataframe as a Scatter Plot
plt.figure(figsize=(10, 6))  # Set the figure size (width, height)
df.plot(kind='scatter' , x='Sales' , y='Profit' , color='green' , marker='x' , s=100)

# Add grid
plt.grid(True , linewidth=0.6 , linestyle=':' , color='blue')

# Add title and axis label
plt.xlabel('Sales')  # Set axis labels
plt.ylabel('Profit')
plt.title('Sales vs. Profit')

# save the plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Pandas_Plots/Scatter_Plot.png', dpi=300, bbox_inches='tight')

# show plot
plt.show()
```

# **5.Line Chart**

```python
# create a dictionary with sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Monthly Sales': [100, 110, 105, 115, 120, 125, 130, 135, 140, 145, 150, 155]
}

# create a pandas dataframe from the dictionary
df = pd.DataFrame(data)

# plot the dataframe as a Line Chart
df.plot(kind='line' , x='Month' , y='Monthly Sales' , marker='o' , linewidth=3 , color='black')

# Add grid
plt.grid(True, linestyle=':', which='major', axis='both' , color='blue')

# Add title and axis label
plt.title('Monthly Sales Trend', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=18 , fontweight='bold')
plt.ylabel('Monthly Sales', fontsize=18 , fontweight='bold')

# improving spacing
plt.tight_layout()

# save the plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Pandas_Plots/Line_Plot.png', dpi=300, bbox_inches='tight')

# show plot
plt.show()
```

# **6.Box Plot**

```python
# Sample data
data = {
    'Product A': [95, 85, 90, 100, 105, 110,150],
    'Product B': [100, 105, 95, 90, 115, 120,155],
    'Product C': [105, 110, 115, 120, 125, 130,160]
}

# Create a pandas dataframe from the dictionary
df = pd.DataFrame(data)

# Customize box appearance
median_prop={'color':'white', 'linewidth':1}
flier_prop={'marker':'^', 'markeredgecolor':'blue', 'markerfacecolor':'green', 'markersize':15}
whisker_prop={'linewidth':2, 'color':'black', 'linestyle':'--'}
box_prop={'color':'black', 'linewidth':2}

# Ploting the Boxplot
df.plot(kind='box' , vert=True , fontsize=16, notch=True, patch_artist=True, showfliers=True,color='black', 
       medianprops=median_prop, flierprops=flier_prop, whiskerprops=whisker_prop, boxprops=box_prop
       )

# Adding gridline
plt.grid(True, alpha=0.5, color='black', linestyle='-', linewidth=1.5)

# Adding title and labels
plt.xlabel('Products', fontsize=16, fontweight='bold')
plt.ylabel('Corresponding Values', fontsize=16, fontweight='bold')
plt.title('Product vs. values', fontsize=18, fontweight='bold')

# Adjust the x-axis ticks
plt.xticks(rotation=45)

# Adjust white spacing
plt.tight_layout()

# save a plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Pandas_Plots/Box_Plot.png', dpi=300, bbox_inches='tight')s

# show plot
plt.show()
```
