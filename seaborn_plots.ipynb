# Line Plot

```python
from matplotlib.font_manager import FontProperties

# Create a plot
ax=sns.lineplot(data=df, x='bill_length_mm', y='bill_depth_mm', hue='sex', style='sex', markers=['s', 'o'], palette='Set1', 
             dashes=False, legend='full')

# Customize title and label
plt.xlabel('bill_length_mm', fontsize=16, fontweight='bold')
plt.ylabel('bill_depth_mm', fontsize=16, fontweight='bold')
plt.title('Line Plot', fontsize=16, fontweight='bold')

# Customize Legend
plt.legend(loc=4, prop=FontProperties(weight='bold', size=16))

# Save the plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Seaborn/Line_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# Heatmap

```python
# Creaet a 2D array
df=np.linspace(0, 10, 20).reshape(4, 5)
```

```python
# Create a plot

# Customization
xlabels=['X1', 'X2', 'X3', 'X4', 'X5']
ylabels=['Y1', 'Y2', 'Y3', 'Y4']
font_annot={'fontsize':12, 'color':'k', 'fontweight':'bold'}
customize_cbar={'shrink':0.6, 'orientation':'horizontal'}

plt.figure(figsize=(8,6)) # Set figure size

x=sns.heatmap(df, annot=True, cbar=True, cmap='Accent', vmin=0, vmax=10,
            xticklabels=xlabels, yticklabels=ylabels, center=0, fmt='.2f', 
            square=True, cbar_kws=customize_cbar, annot_kws=font_annot, 
            linewidths=3, linecolor='lightblue')

# Get the color bar
cbar = x.collections[0].colorbar

# Customize the font properties of the color bar label
cbar.set_label('Color Bar', fontsize=18, fontweight='bold', color='k')

# Customize the font properties of the color bar tick labels
cbar.ax.tick_params(size=10, width=3, color='g', labelsize=12, labelcolor='b')

# rotate x axis label
plt.xticks(rotation=45, fontsize=12, fontweight='bold')

# rotate y axis label
plt.yticks(rotation=45, fontsize=12, fontweight='bold')

# set the title of the plot
plt.title('Heatmap' , fontsize=18 , fontweight='bold')

# Improve spacing
plt.tight_layout() 

# Save a plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Seaborn/Heatmap.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# Pair Plot

```python
# Load Dataset
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
# Create a plot
g=sns.pairplot(df, hue='sex', palette='Dark2', hue_order=['Female', "Male"],      # kind=hist, reg, kde, scatter
             x_vars=['total_bill', 'tip'], kind='scatter', y_vars=['total_bill', 'tip'],  
            diag_kind='hist', markers=['o', 's'], size=3)


# customize legend
g.legend.set_title('Sex', prop=FontProperties(weight='bold', size=16)) # Set Title
g.legend.set_bbox_to_anchor((1, 0.8)) # Set Position
legend_labels = ['Female', 'Male'] # Customize Labels of Legends
for t, l in zip(g._legend.texts, legend_labels):
    t.set_text(l)
    t.set_weight('bold')
    t.set_size(12)


# Set x-axis labels and font properties
for ax in g.axes.flatten():
    ax.set_xlabel(ax.get_xlabel(), fontsize=16, fontweight='bold')

# Set y-axis labels and font properties
for ax in g.axes.flatten():
    ax.set_ylabel(ax.get_ylabel(), fontsize=16, fontweight='bold')

# customize axis tick labels 
for ax in g.axes.flatten():
    ax.tick_params(axis='both', labelsize=12, size=5, width=1.5, 
                   color='g', pad=5, labelcolor='b')     

# Set title
plt.title('Pair Plot', fontsize=22, fontweight='bold', loc='right', x=2.1, y=1.9)

# Save the plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Seaborn/Pair_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# Box Plot

```python
# Load Dataset
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
from matplotlib.font_manager import FontProperties

# Create a plot

# Customize box appearance
median_prop={'color':'w', 'linewidth':2}
flier_prop={'marker':'o', 'markeredgecolor':'k', 'markerfacecolor':'lightblue', 'markersize':15, 'mew':2}
whisker_prop={'linewidth':2, 'color':'blue'}
box_prop={'linewidth':2}
meanprop={'markeredgecolor':'b', 'markeredgewidth':0.3, 'mfc':'k', 'ms':10}
order=['Sun', 'Sat', 'Thur', 'Fri']

plt.figure(figsize=(8, 6)) # set figure size
sns.boxplot(x='day', y='total_bill', data=df, hue='sex', order=order, showmeans=True, palette='viridis', saturation=10, 
             linewidth=2, notch=True, medianprops=median_prop, flierprops=flier_prop, whiskerprops=whisker_prop,
               boxprops=box_prop, meanprops=meanprop)

# Customize Legend
plt.legend(bbox_to_anchor=(1,1), prop=FontProperties(weight='bold', size='16'))

# Adding gridline
plt.grid(alpha=0.8, color='blue', linestyle=':', linewidth=0.5)

# Adding title and labels
plt.xlabel('Day', fontsize=16, fontweight='bold')
plt.ylabel('Total_Bill', fontsize=16, fontweight='bold')
plt.title('Box Plot', fontsize=18, fontweight='bold')

# customize the x-axis ticks
plt.xticks(fontsize=14, fontweight='bold')

# customize the y-axis ticks
plt.yticks(fontsize=14, fontweight='bold')

# Adjust white spacing
plt.tight_layout()

# Save the plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Seaborn/Box_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# Cat Plot

```python
 # Load Dataset
df=sns.load_dataset('tips')
```

```python
# df.head()
df.head()
```

```python
from matplotlib.font_manager import FontProperties

# kind='strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', 'point'

sns.set_style('darkgrid') # set background style

# Create a catplot 
custom_palette = ['b', 'g']
g=sns.catplot(x='day', y='total_bill', data=df, kind='swarm', palette=custom_palette, hue='smoker', 
              saturation=50, orient='v', height=5, aspect=1.5)  

# Set a Line
plt.hlines(y=df['total_bill'].max(), xmin=0, xmax=1.8, color='r', linestyle='-.', label='Max Total Bill', linewidth=2)    

# Remove hue Legend
g.legend.remove()

# Add Legend to show label of line
legend=plt.legend(title='Smoker', prop=FontProperties(weight='bold', size=10), markerscale=2, bbox_to_anchor=(1, 0.6),
           title_fontproperties={'weight':'bold', 'size':16})
# Change the color of the legend title text
legend.get_title().set_color('blue')

# Adding title and labels
plt.xlabel('Day', fontsize=16, fontweight='bold')
plt.ylabel('Total_Bill', fontsize=16, fontweight='bold')
plt.title('Swarm Plot Using Cat Plot', fontsize=18, fontweight='bold')

# customize the x and y axis ticks
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')

# Save the plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Seaborn/Cat_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# Facet Grid

```python
# Load Dataset
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
from matplotlib.font_manager import FontProperties

# Create a multiple plot
fig=sns.FacetGrid(df, col='day', row='smoker', hue='sex', height=3, aspect=1, palette='Dark2')

# map the scatter plot
fig.map(sns.scatterplot,'tip', 'total_bill')

# Customized Legend
legend = plt.legend(prop=FontProperties(weight='bold', size=16), bbox_to_anchor=(1.05, 1), markerscale=2)
legend.set_title('Sex', prop=FontProperties(weight='bold', size=16))
legend.get_title().set_color('red')
# Change color of legend labels
for text in legend.get_texts():
    text.set_color('blue')

# set axis labels
fig.set_axis_labels('Tip_Amount', 'Total_Bill')

# Customize the x and y axis labels, ticks and customized individual subplot title
for ax in fig.axes.flat:
    ax.set_xlabel(ax.get_xlabel(), fontsize=14, fontweight='bold', color='c')
    ax.set_ylabel(ax.get_ylabel(), fontsize=14, fontweight='bold', color='red')

    # Customize the x and y ticks
    ax.set_xticklabels(ax.get_xticks(), fontsize=12, fontweight='bold', color='green')
    ax.set_yticklabels(ax.get_yticks(), fontsize=12, fontweight='bold', color='purple')

    # Customize the titles of the subplots
    title = ax.get_title()
    ax.set_title(title, fontsize=12, fontweight='bold', color='blue')

# Add Title
fig.fig.suptitle('Facet Grid Scatter Plot', y=1.1, fontsize=16, fontweight='bold', color='m')

# Save the plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Seaborn/Facet_Grid.png', dpi=300, bbox_inches='tight')

# Show Plot
plt.show()
```
