# Data Visuilization using **Matplotlib**

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **1-Bar Plot**

```python
x=['Math', 'Science', 'History', 'Civics', 'Pak']
y=[90, 80, 45, 39, 75]

# Create plot
colors=['blue', 'green', 'purple', 'red', 'yellow']
plt.bar(x, y, color=colors, edgecolor='black', linewidth=2, width=0.5, label='Student Performance')

# Add text as a legend
plt.text(2, 90, 'Student Performance', ha='center', va='center_baseline', fontsize=12, fontweight='bold') 

# You can also set legend label using below code in this case you can pass label parameter in plt.bar
# plt.legend(fontsize=15) 

# Add title and labels
plt.xlabel('Subjects', fontsize=16, fontweight='bold')
plt.ylabel('Marks', fontsize=16, fontweight='bold')
plt.title('Pass Card', fontsize=16, fontweight='bold')

# save plot 
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Bar_Plot.png', dpi=300, bbox_inches='tight')

# Display plot
plt.show()
```

# **2-Stack Bar Plot**



```python
x=['Math', 'Science', 'History', 'Civics', 'Pak']
y=[90, 80, 45, 39, 75]
z=[10, 20, 30, 40, 50]
a=[5, 15, 25, 35, 45]

# this code set positions of the bars
t=np.arange(len(x))
t1=[p+0.2 for p in t]
t2=[p+0.4 for p in t]

# Create plot
plt.figure(figsize=(7, 5)) # Set figure size

color1=['blue', 'yellow', 'purple', 'red', 'green']
color2=['green', 'blue', 'yellow', 'purple', 'red']
color3=['m', 'g', 'b', 'c', 'y']

plt.bar(t, y, color=color1, edgecolor='black', linewidth=2, width=0.2, label='Student 1')
plt.bar(t1, z, color=color2, edgecolor='black', linewidth=2, width=0.2, label='Student 2')
plt.bar(t2, a, color=color3, edgecolor='black', linewidth=2, width=0.2, label='Student 3')

# Add legend
plt.legend() 

# Add title and labels
plt.xlabel('Subjects', fontsize=16, fontweight='bold')
plt.ylabel('Marks', fontsize=16, fontweight='bold')
plt.title('Pass Card', fontsize=16, fontweight='bold')

# Center the x-axis label
plt.xticks(t+0.2, x)

# save plot
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Stack_Bar_Plot.png', dpi=300, bbox_inches='tight')

# Display plot
plt.show()
```

# **3-Scatter Plot**

```python
x=[1, 2, 3, 4, 5, 6, 7, 8]
y=[2, 4, 6, 8, 10, 12, 14, 16]
z=[10, 15, 20, 25, 30, 35, 40, 45]

# Apply colorsto dots
colors=['k', 'g', 'y', 'r', 'm', 'b', 'g', 'k']
# increase size of dots
sizes=[50, 80, 110, 140, 180, 210, 240, 270]
color_range=[1, 2, 3, 4, 5, 6, 7, 8]

# Create Plot
plt.figure(figsize=(8, 6)) # set figure size
plt.scatter(x, y, c=colors, s=sizes, marker='^', edgecolors='k')
plt.scatter(x, z, c=color_range, s=sizes, marker='*', edgecolors='k', cmap='Accent')
# make color_bar for star points
color_bar=plt.colorbar()
color_bar.set_label('Color Bar For Star Points', fontsize=16, fontweight='bold')

# Add title and labels
plt.xlabel('x-axis', fontsize=16, fontweight='bold')
plt.ylabel('y-axis', fontsize=16, fontweight='bold')
plt.title('Scatter Plot', fontsize=16, fontweight='bold')

# Save plot
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Scatter_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# **5-Pie Plot**

```python
from matplotlib.font_manager import FontProperties

x=[10, 20, 30, 40, 50]
y=['HTML', 'CSS', 'JS', 'PHP', 'JAVA']

# Create a Plot
plt.figure(figsize=(7,5)) # set figure size
color=['y', 'r', 'm', 'c', 'g']
exp=[0.2, 0, 0, 0, 0] # out the slice form the chart
textprop={'fontsize':16, 'fontweight':'bold', 'color':'blue'}
plt.pie(x, autopct='%0.1f%%', colors=color, labels=y, explode=exp, labeldistance=1.1, textprops=textprop, radius=1.1, 
        startangle=180, counterclock=True, wedgeprops={'linewidth':2, 'edgecolor':'black'}, rotatelabels=0
        )

# You can try both ways to add donut type shape in the pie chart
# plt.pie([1], radius=0.4, colors='white') # way 1
# donut=plt.Circle(xy=(0, 0), radius=0.3, facecolor='white') # way 2
# plt.gca().add_artist(donut)

# Additional customizations
legend_props = FontProperties(weight='bold', size=12)
plt.legend(loc="upper left", title="Programming Languages", bbox_to_anchor=(1.35, 0.75), prop=legend_props)
plt.title("Pie Chart of Programming Languages", fontsize=16, fontweight='bold', x=1.5)

# save the plot
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Pie_Plot.png', dpi=300, bbox_inches='tight')

# Display plot
plt.show()
```

# **6-Stem Plot**

```python
x=[2, 4, 6, 8, 10]
y=[4, 8, 12, 16, 20]

# Create a plot
c=['r', 'g', 'b', 'k', 'm']
plt.stem(x, y, linefmt='k-.', markerfmt='^b', basefmt='g', bottom=1, orientation='vertical', label='Stem PLot')

# legend customization
legend=FontProperties(weight='bold', size=16)
plt.legend(prop=legend, bbox_to_anchor=(0.7, 1))

# Add labels and title
plt.xlabel("X-axis", fontsize=16, fontweight='bold')
plt.ylabel("Y-axis", fontsize=16, fontweight='bold')
plt.title("Python Stem Plot", fontsize=16, fontweight='bold')

# Save Plot
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Stem_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# **7-Box Plot**

```python
x=[10, 20, 30, 40, 50, 60, 70, 80, 90, 150]
x1=[15, 30, 45, 60, 75, 90, 105, 120, 135]

# Create Plot
plt.figure(figsize=(8, 6)) # set fig size

# customize boxplot
boxprop=dict(color='k', linewidth=2, facecolor='#5cb8ad')
medianprop=dict(color='m', linewidth=2)
whiskerprop=dict(color='g', linewidth=2)
capprop=dict(color='r', linewidth=2)
meanprop={'marker': 's', 'markerfacecolor': '#3b3a2f', 'markeredgewidth': 1.5, 'markersize':8, 'markeredgecolor':'k'}
flierprop=dict(marker='o', markerfacecolor='k', markersize=10, markeredgewidth=2, markeredgecolor='m')

# you can also pass a list just put y in the place of x you don't need to write more code
y=[x, x1]

plt.boxplot(x, notch=True, vert=True, widths=0.3, positions=[1], showmeans=True, patch_artist=True, 
             boxprops=boxprop, medianprops=medianprop, whiskerprops=whiskerprop, capprops=capprop, 
             meanprops=meanprop, flierprops=flierprop 
            )

plt.boxplot(x1, notch=True, vert=True, widths=0.3, positions=[2], showmeans=True, patch_artist=True, 
             boxprops=dict(color='b', linewidth=2, facecolor='#4c124d'), medianprops=medianprop, 
             whiskerprops=whiskerprop, capprops=capprop, meanprops=meanprop, flierprops=flierprop 
            )

# Add labels and title
plt.xlabel("X-axis", fontsize=16, fontweight='bold')
plt.ylabel("Y-axis", fontsize=16, fontweight='bold')
plt.title('Box Plot', fontsize=16, fontweight='bold')

# Set x-axis labels
plt.xticks(ticks=[1,2], labels=["Python", "Java"])

# Save Plot
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Box_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# **8-Stack Plot**

```python
from matplotlib.font_manager import FontProperties
x=[1, 2, 3, 4, 5, 6]
area1=[2, 4, 6, 8, 10, 12]
area2=[3, 2, 1, 4, 6, 5]
area3=[7, 9, 6, 5 , 4 ,5]
# Create a plot
area=[area1, area2, area3]
label=['Area 1', 'Area 2', 'Area 3'] 
color=['#350436', '#02071a', '#04d4d4']

plt.stackplot(x, area, labels=label, colors=color, baseline='zero')
# Add legend 
legend_font=FontProperties(weight='bold', size=14)
plt.legend(loc='upper left', prop=legend_font)

# Add labels and title
plt.xlabel("X-axis", fontsize=16, fontweight='bold')
plt.ylabel("Y-axis", fontsize=16, fontweight='bold')
plt.title('Stack Plot', fontsize=16, fontweight='bold')

# Save Plot
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Stack_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# **9-Area Plot**

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [100, 150, 180, 220, 270]  
}

# Create a pandas dataframe from the dictionary
df = pd.DataFrame(data)

# Plotting the area chart
plt.figure(figsize=(10, 6))
sns.set_style("darkgrid")
plt.fill_between(df['Month'], df['Sales'], alpha=0.9, color='black')
plt.plot(df['Month'], df['Sales'], color="darkblue", alpha=0.6, marker='^', markersize=10, linewidth=3)

# Add annotations for specific points
# plt.annotate('Peak Sales', (Month[4], Sales[4]), xytext=(Month[3], Sales[3]),color='green',
#              arrowprops=dict(facecolor='white', shrink=0.05), fontsize=12) 
plt.annotate('Peak Sales', (df['Month'][4], df['Sales'][4]), xytext=(df['Month'][3], df['Sales'][3]),
              color='red',  # Set text color to red
              arrowprops=dict(facecolor='white', shrink=0.05), fontsize=12)


# Add gridlines with different style
plt.grid(True, linestyle=':', linewidth=1, color='darkblue', alpha=0.7)

# Set the title, x-axis and y-axis labels
plt.xlabel('Month', fontsize=14, fontweight='bold')
plt.ylabel('Sales', fontsize=14, fontweight='bold')
plt.title('Monthly Sales Trend', fontsize=16, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Rotate y-axis labels for better readability
plt.yticks(rotation=45)

# adjust whtie spacing
plt.tight_layout()

# Save the plot to a file
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Area_Plot.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

# **10-Step Plot**

```python
from matplotlib.font_manager import FontProperties

x=[1, 2, 3, 4, 5, 6, 7, 8]
y=[2, 4, 6, 8, 10, 12, 14, 16]

# Create a Step Plot
plt.step(x, y, marker='d', ms=10, color='k', mfc='g', label='Python Step Plot', where='mid')

# Add Grid Line
plt.grid(linewidth=0.5, alpha=0.8, linestyle='-.', color='b')

# Add Legend
legend_prop= FontProperties(weight='bold', size=16)
plt.legend(loc='upper left', prop=legend_prop, bbox_to_anchor=(0.025, 0.978))

# Add labels and title
plt.xlabel("X-axis", fontsize=16, fontweight='bold')
plt.ylabel("Y-axis", fontsize=16, fontweight='bold')
plt.title('Step Plot', fontsize=16, fontweight='bold')

# Save Plot
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Step_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# **11-Fill Between Plot**

```python
x=np.array([5, 10, 15, 20, 25, 30, 35, 40])
y=np.array([2, 4, 6, 8, 10, 12, 14, 16])

# create a plot
plt.plot(x, y, color='b')
plt.fill_between(x, y, color='lightgreen', where=(x>=10) & (x<=35), label='Matplotlib Plot')

# Add labels and title
plt.xlabel("X-axis", fontsize=16, fontweight='bold')
plt.ylabel("Y-axis", fontsize=16, fontweight='bold')
plt.title('Fill Between Plot', fontsize=16, fontweight='bold')


# Add Legend
legend_prop= FontProperties(weight='bold', size=16)
plt.legend(loc='upper left', prop=legend_prop)

# Save Plot
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Fill_Between_Plot.png', dpi=300, bbox_inches='tight')

# Display Plot
plt.show()
```

# **12-Subplot**

```python
# create a first plot 
x=[1, 2, 3]
y=[2, 4, 6]
# seperate plot from each other
plt.subplot(2, 2, 1)
plt.plot(x, y, color='r')

# Add Legend
plt.legend(['python'], loc=9, facecolor='k', edgecolor='b', framealpha=0.5, shadow=True)

# create a second plot
x=[2, 3, 4]
y=[8, 10, 12]
# seperate plot from each other
plt.subplot(2, 2, 2)
plt.plot(x, y, color='g')

# Add Legend
plt.legend(['python'], loc=9, facecolor='k', edgecolor='b', framealpha=0.5, shadow=True)

# create a third plot
x=[5, 6, 7]
y=[14, 16, 18]
# seperate plot from each other
plt.subplot(2, 2, 3)
plt.plot(x, y, color='b')

# Add Legend
plt.legend(['python'], loc=9, facecolor='k', edgecolor='b', framealpha=0.5, shadow=True)

# create a fourth plot
x=[8, 9, 10]
y=[20, 22, 24]
# seperate plot from each other
plt.subplot(2, 2, 4)
plt.plot(x, y, color='k')

# Add Legend
plt.legend(['python'], loc=9, facecolor='k', edgecolor='b', framealpha=0.5, shadow=True)

# Save Plot
# plt.savefig('/home/yousuf_shah/Main/Save_Plots/Matplotlib/Subplot.png', dpi=300, bbox_inches='tight',facecolor='lightblue', transparent=False)

# Display Plot
plt.show()
```
